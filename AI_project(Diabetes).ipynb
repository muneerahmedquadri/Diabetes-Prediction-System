```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('Diabetes Classification.csv')
df.head()
```

```python
# Checking the shape of the dataset
df.shape
```

```python
# Checking for null/missing values
df.isnull().sum()
```

```python
# Checking the datatypes
df.dtypes
```

```python
# Unique values in each column
for i in df.columns:
    print(i, df[i].unique())
```

```python
# columns for outlier removal
cols  = ['Age', 'BMI', 'FBS','HbA1c']

#IQR for the selected columns
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

#Threshold for outlier removal
threshold = 1.5

#Find index of outliers
index = np.where((df[cols] < (Q1 - threshold * IQR)) | (df[cols] > (Q3 + threshold * IQR)))[0]

#Drop outliers
df = df.drop(df.index[index])
```

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical variables
cols = ['Age', 'Gender', 'BMI', 'Blood Pressure', 'FBS', 'HbA1c','Family History of Diabetes', 'Smoking', 'Diet', 'Exercise','Diagnosis']

# Label encoding object
le = LabelEncoder()

for i in cols:
    le.fit(df[i])
    df[i] = le.transform(df[i])
    print(i, df[i].unique())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['Diagnosis']), df['Diagnosis'], test_size = 0.2, random_state = 0)
```

## Diabetes Disease Prediction

For predicting the diabetes disease, I have used the following classification models:
1. Random Forest Classifier
2. Decision Tree Classifier

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest object
rfc = RandomForestClassifier(random_state=0, max_features='sqrt', n_estimators=200, class_weight='balanced')
```

```python
# Training the model
rfc.fit(X_train, y_train)
```

```python
# Training accuracy
rfc.score(X_train, y_train)*100
```

```python
# Predicting the test set results
rfc_pred = rfc.predict(X_test)
```

 ### Decision Tree Classifier

```python
 from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree object
dtc = DecisionTreeClassifier(random_state=0, max_depth= 12, min_samples_leaf=2, min_samples_split=2, class_weight='balanced')
```

```python
# Training the model
dtc.fit(X_train, y_train)
```

```python
# Training accuracy
dtc.score(X_train, y_train)*100
```

```python
# Predicting the test set results
dtc_pred = dtc.predict(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
#Training the model
lr.fit(X_train, y_train)
```

```python
#Training accuracy
lr.score(X_train, y_train)*100
```

```python
#Predicting the test set results
lr_pred = lr.predict(X_test)
```

## Model Evalution

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
fig, ax  = plt.subplots(1,3, figsize = (20,5))
sns.heatmap(confusion_matrix(y_test, rfc_pred), annot = True, cmap = 'coolwarm', ax = ax[0]).set_title('Random Forest')
sns.heatmap(confusion_matrix(y_test, dtc_pred), annot = True, cmap = 'coolwarm', ax = ax[1]).set_title('Decision Tree')
sns.heatmap(confusion_matrix(y_test, lr_pred), annot = True, cmap = 'coolwarm', ax = ax[2]).set_title('Logistic Regression')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Random Forest')
print('Accuracy Score: ', accuracy_score(y_test, rfc_pred)*100)
print('Precision Score: ', precision_score(y_test, rfc_pred)*100)
print('Recall Score: ', recall_score(y_test, rfc_pred)*100)
print('F1 Score: ', f1_score(y_test, rfc_pred)*100)
```

```python
print('Decision Tree')
print('Accuracy Score: ', accuracy_score(y_test, dtc_pred)*100)
print('Precision Score: ', precision_score(y_test, dtc_pred)*100)
print('Recall Score: ', recall_score(y_test, dtc_pred)*100)
print('F1 Score: ', f1_score(y_test, dtc_pred)*100)
```

```python
print('Logistic Regression')
print('Accuracy Score: ', accuracy_score(y_test, lr_pred)*100)
print('Precision Score: ', precision_score(y_test, lr_pred)*100)
print('Recall Score: ', recall_score(y_test, lr_pred)*100)
print('F1 Score: ', f1_score(y_test, lr_pred)*100)
```
